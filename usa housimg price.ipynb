```python
from sklearn import preprocessing
label_encoding=preprocessing.LabelEncoder()
label_encoding.fit(df['Address'])
df['Address']=label_encoding.transform(df['Address'].astype(str))
df.head()
```

```python
df.columns
```

```python
X=df.drop("Price",axis=1)
Y=df["Price"]
X.shape  , Y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
from sklearn.linear_model import LinearRegression
model=LinearRegression(normalize=True).fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print("Training score : ",model.score(x_train,y_train))
print("Testing score : ",r2_score(y_test,y_pred))
```

```python
new_model=BaggingRegressor(DecisionTreeRegressor(),n_estimators=500,bootstrap=False,max_samples=1.0,n_jobs=-1)
new_model.fit(x_train,y_train)
new_y_pred=new_model.predict(x_test)
from sklearn.metrics import r2_score
print("Training score : ",model.score(x_train,y_train))
print("Testing score : ",r2_score(y_test,new_y_pred))
```
